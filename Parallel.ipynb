```python
import datetime


from PIL import Image, ImageOps
from astropy.io import fits
import pyvips
import numpy as np

def convert_ccd(exp):
    start = datetime.datetime.now()
    name=exp.split('det')[1].split('.fits')[0]
    ccd = fits.open(exp, memmap=True)
    imgccd = ccd[1].data

    # Add border
    pil_img=Image.fromarray(imgccd)
    img_with_border = ImageOps.expand(pil_img,border=25,fill='grey')
    # Save tiff
    img_arr = np.array(img_with_border.convert('L'))
    vips=pyvips.Image.new_from_memory(img_arr.data, img_arr.shape[1], img_arr.shape[0], 1, 'uchar')
    outputdir='./expsNew'

    vips.tiffsave("{}/{}.tif".format(outputdir, name), tile=True, tile_width=256, tile_height=256, pyramid=True, bigtiff=True, compression='deflate')
    
    end = datetime.datetime.now()
    return 'DONE {}: {}'.format(name, str(end-start))
```

```python
import datetime
import glob
import concurrent.futures

exps = glob.glob(
    '/home/felipe/repos/lsst/lsst_image/data_repo/calexp/00193824-g/R*/calexp*')

start = datetime.datetime.now()

with concurrent.futures.ProcessPoolExecutor() as executor:
     for time in executor.map(convert_ccd, exps):
         print(time)

end = datetime.datetime.now()
print('DONE: {}'.format(str(end-start)))
```
